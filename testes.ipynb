```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('pizzas.csv')
```

```python
df
```

```python
df.plot(kind='scatter', x='diametro', y='preco', title='Correlação entre Diâmetro e Preço');
```

```python
from sklearn import linear_model

modelo = linear_model.LinearRegression()
x = df[["diametro"]]
y = df[["preco"]]

modelo.fit(x, y)
```

```python
print(modelo.predict([[30]])[0][0])
```

```python
print(modelo.predict([[110]])[0][0])
```
